```python
def daily_mape(original,predict,method):
    filename = filename_pre+method+'.xlsx'
    writer = pd.ExcelWriter(filename)

    df_original = pd.DataFrame(original,columns=node_list)
    df_original = df_original.loc[:,select_node_list]
    df_original.loc[:,'day'] = day
    df_original.loc[:,'hour'] = hour
    df_original.loc[:,'weekday'] = date['weekday'].values
    df_original_daily = pd.pivot_table(df_original,index='day',aggfunc='sum')
    
    df_predict = pd.DataFrame(predict,columns=node_list)
    df_predict = df_predict.loc[:,select_node_list]
    df_predict.loc[:,'day'] = day
    df_predict.loc[:,'hour'] = hour
    df_predict.loc[:,'weekday'] = date['weekday'].values
    df_predict_daily = pd.pivot_table(df_predict,index='day',aggfunc='sum')
    
#####
    # Daily MAPE for each nodes for each day
    daily_MAPE = abs((df_predict_daily - df_original_daily)/df_original_daily).iloc[:,0:len(select_node_list)]
    
    daily_MAPE_f = daily_MAPE.iloc[:,0:1].copy()
    daily_MAPE_f.loc[:,'node'] = daily_MAPE_f.columns.values[0]    
    daily_MAPE_f.columns = ['MAPE','node']
    
    # Transform Data so that daily_MAPE_f have a ['node'] columnns 
    for i in range(1,len(select_node_list)):
        select = daily_MAPE.iloc[:,i:(i+1)].copy()
        select.loc[:,'node'] = select.columns.values[0]
        select.columns = ['MAPE','node']
        daily_MAPE_f = pd.concat([daily_MAPE_f,select])    
    
    # Transform Data so that daily_MAPE_f have a ['type'] columns and output
    daily_MAPE_f.loc[:,'day'] = daily_MAPE_f.index
    daily_MAPE_f= pd.merge(daily_MAPE_f, node_info_type, on='node')
    daily_MAPE_f = daily_MAPE_f[['node','day','type','MAPE']]
    #print(daily_MAPE_f.head())
    #daily_MAPE_f.to_excel(writer,'Daily')
    daily_results = copy.deepcopy(daily_MAPE_f)
    
    
######
    daily_MSE = ((df_predict_daily - df_original_daily)**2).iloc[:,0:len(select_node_list)]
    daily_MSE_f = daily_MSE.iloc[:,0:1].copy()
    daily_MSE_f.loc[:,'node'] = daily_MSE_f.columns.values[0]
    daily_MSE_f.columns = ['MSE','node']
    
    # Transform Data so that daily_MSE_f have a ['node'] columnns 
    for i in range(1,len(select_node_list)):
        select = daily_MSE.iloc[:,i:(i+1)].copy()
        select.loc[:,'node'] = select.columns.values[0]
        select.columns = ['MSE','node']
        #print(select)
        #select
        daily_MSE_f = pd.concat([daily_MSE_f,select])    
    
    # Transform Data so that daily_MSE_f have a ['type'] columns and output
    daily_MSE_f['day'] = daily_MSE_f.index
    daily_MSE_f= pd.merge(daily_MSE_f, node_info_type, on='node')
    daily_MSE_f = daily_MSE_f[['node','day','type','MSE']]
    #print(daily_MSE_f.head())
    daily_results = pd.merge(daily_results, daily_MSE_f, on= ['node','day','type'])
    
#####
    daily_nMSE = ((df_predict_daily - df_original_daily)**2).iloc[:,0:len(select_node_list)]/((df_original_daily.iloc[:,0:len(select_node_list)].max() -df_original_daily.iloc[:,0:len(select_node_list)].min())**2)
    daily_nMSE_f = daily_nMSE.iloc[:,0:1].copy()
    daily_nMSE_f.loc[:,'node'] = daily_nMSE_f.columns.values[0]
    daily_nMSE_f.columns = ['nMSE','node']
    
    # Transform Data so that daily_nMSE_f have a ['node'] columnns 
    for i in range(1,len(select_node_list)):
        select = daily_nMSE.iloc[:,i:(i+1)].copy()
        select.loc[:,'node'] = select.columns.values[0]
        select.columns = ['nMSE','node']
        #print(select)
        #select
        daily_nMSE_f = pd.concat([daily_nMSE_f,select])    
    
    # Transform Data so that daily_nMSE_f have a ['type'] columns and output
    daily_nMSE_f['day'] = daily_nMSE_f.index
    daily_nMSE_f= pd.merge(daily_nMSE_f, node_info_type, on='node')
    daily_nMSE_f = daily_nMSE_f[['node','day','type','nMSE']]
    #print(daily_nMSE_f.head())
    daily_results = pd.merge(daily_results, daily_nMSE_f, on= ['node','day','type'])
    
    
######
    daily_MAE = abs(df_predict_daily - df_original_daily).iloc[:,0:len(select_node_list)]
    daily_MAE_f = daily_MAE.iloc[:,0:1].copy()
    daily_MAE_f.loc[:,'node']= daily_MAE_f.columns.values[0]
    daily_MAE_f.columns = ['MAE','node']

    # Transform Data so that daily_MAE_f have a ['node'] columnns 
    for i in range(1,len(select_node_list)):
        select = daily_MAE.iloc[:,i:(i+1)].copy()
        select.loc[:,'node'] = select.columns.values[0]
        select.columns = ['MAE','node']
        #print(select)
        #select
        daily_MAE_f = pd.concat([daily_MAE_f,select])    
    
    # Transform Data so that daily_MAE_f have a ['type'] columns and output
    daily_MAE_f['day'] = daily_MAE_f.index
    daily_MAE_f= pd.merge(daily_MAE_f, node_info_type, on='node')
    daily_MAE_f = daily_MAE_f[['node','day','type','MAE']]
    #print(daily_MAE_f.head())
    daily_results = pd.merge(daily_results, daily_MAE_f, on= ['node','day','type'])
    
#####
    daily_MARNE = (abs(df_predict_daily - df_original_daily).iloc[:,0:len(select_node_list)])/(df_original_daily.iloc[:,0:len(select_node_list)].abs().max())
    daily_MARNE_f = daily_MARNE.iloc[:,0:1].copy()
    daily_MARNE_f.loc[:,'node'] = daily_MARNE_f.columns.values[0]
    daily_MARNE_f.columns = ['MARNE','node']

    # Transform Data so that daily_MARNE_f have a ['node'] columnns 
    for i in range(1,len(select_node_list)):
        select = daily_MARNE.iloc[:,i:(i+1)].copy()
        select.loc[:,'node'] = select.columns.values[0]
        select.columns = ['MARNE','node']
        #print(select)
        #select
        daily_MARNE_f = pd.concat([daily_MARNE_f,select])    
    
    # Transform Data so that daily_MARNE_f have a ['type'] columns and output
    daily_MARNE_f['day'] = daily_MARNE_f.index
    daily_MARNE_f= pd.merge(daily_MARNE_f, node_info_type, on='node')
    daily_MARNE_f = daily_MARNE_f[['node','day','type','MARNE']]
    #print(daily_MARNE_f.head())
    daily_results = pd.merge(daily_results, daily_MARNE_f, on= ['node','day','type'])

#####
    daily_results['nRMSE'] = daily_results['nMSE'].apply(np.sqrt)
    daily_results['RMSE'] = daily_results['MSE'].apply(np.sqrt)
    daily_results = daily_results[['node','day','type','MAPE','MSE','nMSE','RMSE','nRMSE','MAE','MARNE']]
    daily_results.to_excel(writer,'Daily')
    
    daily_results_node = pd.pivot_table(daily_results,index ='node',values = ['MAE','MARNE','MSE','nMSE','MAPE'])
    daily_results_node['nRMSE'] = daily_results_node['nMSE'].apply(np.sqrt)
    daily_results_node['RMSE'] = daily_results_node['MSE'].apply(np.sqrt)
    daily_results_node['node'] = daily_results_node.index
    daily_results_node.index.name = ""
    daily_results_node= pd.merge(daily_results_node, node_info_type, on='node')
    daily_results_node = daily_results_node[['node','type','MAPE','MSE','nMSE','RMSE','nRMSE','MAE','MARNE']]                                        
    daily_results_node.to_excel(writer,'Daily_node_average')

    daily_results_type = pd.pivot_table(daily_results_node,index = 'type', values = ['MAPE','MSE','nMSE','RMSE','nRMSE','MAE','MARNE'])
    #daily_results_type = pd.pivot_table(daily_results,index ='type',values = ['MAE','MARNE','MSE','nMSE','MAPE'])
    #daily_results_type['nRMSE'] = daily_results_type['nMSE'].apply(np.sqrt)
    daily_results_type = daily_results_type[['MAPE','MSE','nMSE','RMSE','nRMSE','MAE','MARNE']]
    daily_results_type.to_excel(writer,'Daily_type_average')
    
    writer.save()
    
#     plt.hist(daily_MAPE_f['MAPE'],range=(0,0.6))
#     plt.show();
#     plt.hist(pd.pivot_table(daily_MAPE_f,index ='node')['MAPE'].values)
#     plt.show();
```

```python
def comparemodels(criterion,frequency):
    alternative_method_list = ['FARLSTM','FAR','CNN1D','LSTM','MLP','AR','SAR','VAR','lightGBM']
    method = 'FARCNN'
    level = 'node' #node/type
    filename = filename_pre + method + '.xlsx'
    sheetname  = frequency + '_' + level + '_average'
    df = pd.read_excel(filename,header=0,sheet_name=sheetname)
    results = df.loc[:,['Unnamed: 0','node','type',criterion]]
    results.columns = ['Node Index','node','type',method]
    for method in alternative_method_list:
        level = 'node' #node/type
        filename = filename_pre + method+'.xlsx'
        sheetname  = frequency + '_' + level + '_average'
        df = pd.read_excel(filename,header=0,sheet_name=sheetname)
        df = df.loc[:,['Unnamed: 0','node','type',criterion]]
        df.columns = ['Node Index','node','type',method]
        df[method] = df[method]/results['FARCNN']-1
        results = pd.merge(results,df,on=['Node Index','node','type'])
    results.iloc[:,3:] = results.iloc[:,3:]*100
    results_type = pd.pivot_table(results,index =['type'],values = ['FARCNN','FARLSTM','FAR','CNN1D','LSTM','MLP','AR','SAR','VAR','lightGBM'], margins= True)
    results_type = results_type[['FARCNN','FARLSTM','FAR','CNN1D','LSTM','MLP','AR','SAR','VAR','lightGBM']]
    
    outputfile = filename_pre +  'ResultsSummary.csv'
#     book = load_workbook(outputfile)
#     writer = pd.ExcelWriter(outputfile,engine='openpyxl')
#     writer.book = book
#     sheetname = criterion+frequency
#     results_type.to_excel(writer,sheetname)     
    with open(outputfile,'a') as fd:
        results_type.to_csv(fd)
    return results_type
```
