```python
import pandas as pd
import numpy as np
```

```python
data_total = pd.read_csv("newdata/Gas_flow_19.04.16-16.01.18_200.csv",header = 0, sep = ';')
```

```python
data_total.head()
```

```python
node_sub = list(np.unique(data_total['node']))
```

```python
len(node_sub)
```

```python
node_sub
```

```python
data_group =data_total.groupby('node')
```

```python
data_group.count()
```

```python
for i in node_sub:
    path = 'splitdata/'+ i +'.csv'
    data_total[data_total['node']== i].to_csv(path)
```

```python
for i in node_sub:
    path = 'reshapedata/'+ i +'.csv'
    dataset = data_total[data_total['node']== i]
    
    flow = dataset['flow'].values

    flow_reshape = flow.reshape((len(flow)//24,24))

    flow_reshape = flow_reshape.T

    with open(path, 'ab') as f1:
        np.savetxt(f1, flow_reshape,delimiter=",") 
```

```python
with open("nodes_list.csv", "w") as f:
    for s in node_sub:
        f.write(str(s) +"\n")
```
